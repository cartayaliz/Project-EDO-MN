```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------
# EDO: dz/dt = mu + z - z^2
# Puntos de Equilibrio: z = (1  sqrt(1 + 4*mu)) / 2
# Condicin: mu >= -0.25
# ------------------------------------------------

# 1. Definir el rango del parmetro de bifurcacin (mu)
mu_critico = -0.25
mu_max = 2.0
num_puntos = 500

# Creamos un vector de mu, asegurando que mu >= mu_critico
mu_values = np.linspace(mu_critico, mu_max, num_puntos)

# 2. Calcular los dos puntos de equilibrio z* para cada mu
# Discriminante: Delta = 1 + 4*mu
delta = 1 + 4 * mu_values

# Rama Inferior (Inestable): z1* = (1 - sqrt(Delta)) / 2
z1_inestable = (1 - np.sqrt(delta)) / 2

# Rama Superior (Estable): z2* = (1 + sqrt(Delta)) / 2
z2_estable = (1 + np.sqrt(delta)) / 2

# 3. Construir el Diagrama de Bifurcacin
plt.figure(figsize=(10, 6))

# Trazar la rama estable (lnea continua)
plt.plot(mu_values, z2_estable, 'b-', label='Estable ($z_2^*$)', linewidth=2)

# Trazar la rama inestable (lnea punteada)
plt.plot(mu_values, z1_inestable, 'b--', label='Inestable ($z_1^*$)', linewidth=2)

# 4. Resaltar el Punto de Bifurcacin
# El punto de bifurcacin es (mu_c, z_c) = (-0.25, 0.5)
plt.plot(mu_critico, 0.5, 'ro', markersize=8, label='Punto de Bifurcacin', zorder=5)
plt.axvline(mu_critico, color='r', linestyle=':', alpha=0.6) # Lnea vertical en mu_c

# 5. Configuracin y Etiquetas del Grfico
plt.title('Diagrama de Bifurcacin: $\\frac{dz}{dt} = \\mu + z - z^2$', fontsize=16)
plt.xlabel('Parmetro de Diseo $\\mu$', fontsize=14)
plt.ylabel('Puntos de Equilibrio $z^*$ (Razn de Vaciado)', fontsize=14)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(loc='lower right', fontsize=12)

# Aadir texto para identificar el tipo de bifurcacin
plt.text(mu_critico + 0.1, 1.5, 'Bifurcacin Silla-Nodo',
         color='k', fontsize=12, bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

plt.ylim(np.min(z1_inestable) - 0.1, np.max(z2_estable) + 0.1)
plt.xlim(mu_critico - 0.2, mu_max)

plt.show()
```
